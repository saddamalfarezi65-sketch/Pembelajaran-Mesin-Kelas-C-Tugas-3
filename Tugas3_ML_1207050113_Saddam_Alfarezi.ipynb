```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
baca = pd.read_csv('/jumlah_produksi_sampah_menurut_jenisnya_di_kota_ban_2.csv')
baca.head()
```

```python
baca.info()
```

```python
baca.head()
```

```python
#Menentukan variabel yang akan di klusterkan
baca_x = baca[['produksi_sampah', 'tahun']]
baca_x.head()
#Memvisualkan persebaran data
plt.scatter(baca.produksi_sampah, baca.tahun, s =10, c = "red", marker = "o", alpha = 0.5)
plt.show()
```

```python
x_array = np.array(baca_x)
print(x_array)

```

```python
x_array = np.array(baca_x) # Ensure x_array is updated from the correct baca_x
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans.fit(x_scaled)
```

```python
print(kmeans.labels_)
```

```python
baca["kluster"] = kmeans.labels_
```

```python
print(baca.head())
```

```python
# These metrics (accuracy_score, classification_report, f1_score) are typically used
# for evaluating supervised classification models, which is not applicable here as
# K-Means is an unsupervised clustering algorithm and does not have 'true' labels (y_test)
# to compare against.
# You should consider using appropriate metrics for clustering evaluation like silhouette score or inertia.
# For now, this code has been commented out as it is not suitable for this task.
```

```python
import matplotlib.pyplot as plt

# Plot hasil clustering
plt.figure(figsize=(8,6))

plt.scatter(
    baca['produksi_sampah'],
    baca['tahun'],
    c=baca['kluster'],
    cmap='viridis'
)

# Titik centroid
plt.scatter(
    kmeans.cluster_centers_[:, 0],   # produksi_sampah (index 0 in x_scaled)
    kmeans.cluster_centers_[:, 1],   # tahun (index 1 in x_scaled)
    s=200,
    c='red',
    marker='X',
    label='Centroid'
)

plt.xlabel('Produksi Sampah')
plt.ylabel('Tahun')
plt.title('K-Means Clustering')
plt.legend()

plt.show()
```
